````python
import requests

resp=requests.get('https://www.sohu.com/')
if resp.status_code==200:
  print(resp.text)
````

````python
import re

pattern=re.compile(r'<a.*?href="(.*?)".*?title="(.*?)".*?>')
resp=requests.get('http://www.sohu.com/')
if resp.status_code==200:
  all_matches=pattern.findall(resp.text)
  for href,title in all_matches:
    print(href)
    print(title)
````
