#### 绘制郑州中午11点-12点温度变化趋势图

```python
from pylab import mpl
```

```python
# # 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# # 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False

# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)
x = list(range(60))
y = [random.randint(15, 26) for i in x]
# 2.绘制折线图
plt.plot(x, y, label="郑州")
# 设置标题
plt.title("郑州中午11点-12点温度变化趋势图", fontsize=24)
# 设置xy轴的标题
plt.xlabel("时间（min）", fontsize=16)
plt.ylabel("温度（℃）", fontsize=16)
# 设置y轴刻度的取值范围
plt.ylim(0, 30)
# 设置x轴刻度的显示内容
x_ticks = x[::5]
x_ticks_label = [f"11点{i}分" for i in x_ticks]
plt.xticks(ticks=x_ticks, labels=x_ticks_label)
# 绘制坐标系中的网格
plt.grid(linestyle='--',alpha=0.8)
# 设置图例
plt.legend(loc=0)
# 3.显示图形
# 保存图形
plt.savefig("./test01.png")
plt.show()
```

#### 绘制郑州开封两座城市中午11点-12点的温度变化趋势图

```python
# # 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# # 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False

# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)
x = list(range(60))
y = [random.randint(20, 28) for i in x]
y_kaifeng = [random.randint(18, 23) for i in x]
# 2.绘制折线图
plt.plot(x, y, label="郑州")
plt.plot(x, y_kaifeng, label='开封',linestyle='--', color='r')


# 设置标题
plt.title("郑州开封中午11点-12点温度变化趋势图", fontsize=24)
# 设置xy轴的标题
plt.xlabel("时间（min）", fontsize=16)
plt.ylabel("温度（℃）", fontsize=16)
# 设置y轴刻度的取值范围
plt.ylim(0, 30)
# 设置x轴刻度的显示内容
x_ticks = x[::5]
x_ticks_label = [f"11点{i}分" for i in x_ticks]
plt.xticks(ticks=x_ticks, labels=x_ticks_label)
# 绘制坐标系中的网格
plt.grid(linestyle='--',alpha=0.8)
# 设置图例
plt.legend(loc=0)
# 3.显示图形
# 保存图形
plt.savefig("./test01.png")
plt.show()
```

#### 同一画布绘制不同坐标系绘制图形

```python
# 绘制画布
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)

# 向子坐标系中绘制图形
x = list(range(60))
y = [random.randint(20, 28) for i in x]
y_kaifeng = [random.randint(18, 23) for i in x]
# 绘制两个坐标系的折线图
axes[0].plot(x, y, label="郑州")
axes[1].plot(x, y_kaifeng, color='r', linestyle='--', label="开封")
# 坐标系标题
axes[0].set_title("郑州中午11点到12点温度变化趋势图", fontsize=20)
axes[1].set_title("开封中午11点到12点温度变化趋势图", fontsize=20)
# 设置y轴刻度的取值范围
axes[0].set_ylim(0, 30)
axes[1].set_ylim(0, 30)
# 设置网格
axes[0].grid()
axes[1].grid(linestyle="--", alpha=0.6)
# 设置图例
axes[0].legend()
axes[1].legend()

plt.savefig("./test02.png")
plt.show()
```
